# 3 Controladores en el dominio del tiempo

## 3.1 Introducción
Para iniciar el proceso de diseño de un sistema de control automático, es necesario considerar los siguientes puntos:

- Tener pleno conocimiento de lo que debe hacer el sistema y como hacerlo (especificaciones del diseño).
- Seleccione el mejor controlador para el proceso. 
- Determine los valores de los parámetros del controlador para alcanzar los objetivos de diseño.



## 3.2 Características de los controladores

A continuacion se mencionan algunas consideraciones que se deben tomar en cuenta al realizar un controlador en el dominio del tiempo.

- El PID es uno de los controladores más ampliamente usado en los esquemas de compensación, se diseña en el dominio del tiempo.
		
- Los polos complejos conjugados producen una respuesta subamortiguada. Si los polos son reales, la respuesta al escalón es sobreamortiguada; sin embargo los ceros en lazo cerrado pueden causar sobretiros.
		
- La respuesta de un sistema está dominada por aquellos polos más cercanos al origen del plano s.
		
- Mientras más alejados a la izquierda del plano s se ubiquen los polos dominantes del sistema, el sistema responderá más rápido, será mayor su ancho de banda, se incrementará su costo de implementación debido a que las señales internas aumentan.
	 
- Cuando un polo y un cero, de una función de transferencia de un sistema, se cancela uno con el otro, la porción de la respuesta del sistema asociada con el polo tendrá una magnitud más pequeña.

## 3.4 Controlador PI
 
* Añade un cero en s=-$k_i/k_p$ a la función de transferencia de la trayectoria directa.
* Añade un polo en s=0 a la función de transferencia de la trayectoria directa, de manera que sí el error en estado estable es constante lo reduce a cero.
* Un método para diseñar un PI es seleccionar el cero en s=$-k_i/k_p$ cerca del origen y lejos de los polos significativos y los valores de $k_i$ y $k_p$ deben ser pequeños.
